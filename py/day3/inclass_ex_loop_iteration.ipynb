```python
# for loop example
for x in [10, 20, 30]:
    print(x)
```

```python
# While loop example
count = 1
while count <= 3:
    print(count)
    count += 1
```

```python
# Nested loops example
for row in range(1, 4):        # Outer loop
    for col in range(1, 3):    # Inner loop
        print(f"Row {row}, Col {col}")
```

Iteration process:

    Outer loop starts with row = 1

        Inner loop runs fully: col = 1, col = 2

    Outer loop moves to row = 2

        Inner loop runs fully again: col = 1, col = 2

    Outer loop moves to row = 3

        Inner loop runs fully again: col = 1, col = 2

    Outer loop ends

ðŸ‘‰ Each outer iteration triggers all inner iterations.
